```python
import pandas as pd
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
export_2015 = pd.read_excel('Russia2015.xlsx', sheet_name = 'Partner')
export_2016 = pd.read_excel('Russia2016.xlsx', sheet_name = 'Partner')
```

```python
export = pd.concat([export_2015, export_2016])
```

```python
export.sample(10)
```

```python
exclude = ['Europe & Central Asia', ' World', 'East Asia & Pacific', 'Unspecified', 'Middle East & North Africa']
export_filterd = export[-export['Partner Name'].isin(exclude)]
```

```python
export_pivot = export_filterd.pivot_table(index = ['Year','Partner Name'], values = 'Export (US$ Thousand)', aggfunc='sum')\
    .sort_values(['Year','Export (US$ Thousand)'], ascending= [True, False])
```

```python
export_pivot[export_pivot.index.get_level_values('Year') == 2015]
```
