```python
streets = {}
```


Считает колличесто остановок на одной улице

```python
%time
with open('bus.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f, delimiter=";")
    for row in reader:
        if row["Street"] in streets:
            streets[row['Street']] += 1
        else:
            streets[row['Street']] = 1        
```

```python
%time
sorted(streets.items(), key= lambda item: item[1], reverse = True)
```

```python
RepairOfEscalators = {}
```

Считает колличество ремонтов эксковаторов в метро

```python
with open('metro.csv', 'r', encoding = 'cp1251') as f:
    reader = csv.DictReader(f, delimiter=';')
    for row in reader:
        if row['RepairOfEscalators'] != '' and row['RepairOfEscalators'] in RepairOfEscalators:
            RepairOfEscalators[row['Name']] += 1
        elif row['RepairOfEscalators'] != '' and row['RepairOfEscalators'] not in RepairOfEscalators:
            RepairOfEscalators[row['Name']] = 1
        else:
            pass
            
```

```python
sorted(RepairOfEscalators.items(), key= lambda item: item[1], reverse = True)
```

```python
from geopy.distance import great_circle
```

```python
metro_a_bus = {}
bus_geo_station = {}
metro_geo_data = {}
```

```python
with open('metro.csv', 'r', encoding = 'cp1251') as f:
    reader = csv.DictReader(f, delimiter=';')
    for row in reader:
        if row["Name"] not in metro_geo_data:
            metro_geo_data[row["Name"]] = [row["Longitude_WGS84"], row["Latitude_WGS84"]]  
```

```python
with open('bus.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f, delimiter=";")
    for row in reader:
        bus_geo_station[row['global_id']]= [row["Longitude_WGS84"], row["Latitude_WGS84"]]
```

```python
bus_geo_station = sorted(bus_geo_station.items(), key= lambda item: item[1], reverse = True)[1:]
bus_geo_station
```

```python
metro_geo_data = sorted(metro_geo_data.items(), key= lambda item: item[1], reverse = True)[1:]
metro_geo_data
```

```python
for metro in metro_geo_data:
    num1, num2 = metro[1]
    coordinate1 = num1, num2
    for bus in bus_geo_station:
        num1, num2 = bus[1]
        coordinate2 = num1, num2
        distance = great_circle(coordinate1, coordinate2).km
        if distance <= 0.5 and metro_a_bus.get(metro[0]) != None:
            metro_a_bus[metro[0]] += 1
        elif distance <= 0.5 and metro_a_bus.get(metro[0]) == None:
            metro_a_bus[metro[0]] = 1        
```
