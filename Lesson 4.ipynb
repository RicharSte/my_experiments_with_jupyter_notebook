```python
import random
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap
```

```python
data_x1 = range(0, 100, 5)
data_y1 = range(0,20)
plt.plot(data_x1, data_y1)
plt.show()
```

```python
data_x2 = random.sample(range(500), 10)
data_y2 = random.sample(range(100), 10)
data_x2.sort()
data_y2.sort()
plt.plot(data_x2, data_y2)
plt.show()
```

```python
plt.plot(data_x1, data_y1, color='red', linestyle='dashed')
plt.figure()
plt.plot(data_x2, data_y2, color='green', linestyle='dotted')
plt.show()
```

```python
stations = pd.read_csv(open('bus.csv', 'r', encoding='cp1251'), sep=';')
stations
```

```python
pavilion_count = stations.groupby(['Pavilion']).size()
```

```python
pavilion_count
```

```python
plt.pie(pavilion_count, labels=pavilion_count.index, autopct='%.2f')
plt.title("Павильоны на остановке")
plt.legend()
plt.show()
```
